## **Executive Summary**

Malaria is an extremely dangerous and potentially fatal disease transmitted to humans by a peripheral blood parasite of the genus Plasmodium. The disease generally occurs in tropical countries. In 2022, WHO recorded more than 249 million cases of malaria and 608,000 malaria-related deaths in 85 countries, with more than 90% of these cases and deaths occurring in the WHO African region. 80% of deaths in this region are children under the age of 5. Despite its dangerous nature, however, malaria is curable. A WHO goal for malaria is elimination in at least 35 countries by 2030. But before elimination can be achieved, Accurate and timely detection of malaria is crucial for effective treatment and control of the disease, particularly in regions where it is endemic.

Despite advancements in newer techniques, manual microscopy for examining blood smears—developed in the late 19th century—remains the most widely used method for diagnosing malaria. However, this process requires specialized training and considerable expertise to ensure accuracy. Several field studies have shown that when performed by non-experts, particularly in rural, malaria-endemic areas, manual microscopy often proves unreliable due to inadequate training. Therefore, we believe that an automated malaria detection system could significantly enhance disease detection and contribute to its eventual elimination. Such a system would eliminate human error and be calibrated for maximum precision, thereby minimizing diagnostic mistakes.

The goal of this project is to develop a robust and efficient Convolutional Neural Network (CNN) model for the detection of malaria-infected cells from microscopic images. This model aims to assist healthcare professionals in diagnosing malaria more quickly and accurately, ultimately improving patient outcomes and reducing the burden on medical facilities. Multiple model architectures and techniques such as conversion from RGB to HSV, Gaussian blurring and Data Augmentation have been explored. Finally, the model adopted is the model with data augmentation. By leveraging data augmentation techniques such as horizontal flip, zoom range, shear range and rotation, the goal was to enhance the model's ability to generalize and accurately classify images as either parasitized or uninfected.

### **Key findings**

- Model accuracy: The CNN model adopted achieved an impressive accuracy of 97%.

- Recall for Parasitized cells: The model demonstrated a high recall rate of 98% for detecting parasitized cells, indicating its strong ability to correctly identify infected samples.

- Recall for Uninfected cells: The model also achieved a 97% recall for uninfected cells, ensuring reliable detection of healthy samples.

- Data Augmentation: By employing data augmentation techniques, the model's robustness was significantly improved, enabling it to handle a wider variety of image variations and conditions.

### **Specifications of the final model**

The final model with data augmentation is made up of 5 convolutional blocks and 2 dense blocks, both types of block having a “relu” activation function. A convolutional block consists of a Conv2D layer, batch normalization, Maxpooling and Dropout. A dense block consists of a dense layer, batch normalization and dropout. We end with an output layer consisting of 2 units and a “softmax” activation function. We achieved an overall accuracy of 97%. The accuracy graph shows a fairly good match between training and validation.

```python
# Creating a variable to store the path of the extracted train dataset
train_dir = '/content/drive/MyDrive/Learning/ADSP/Capstone_Project/datasets/cell_images/train'

# Designed size so all the images have the same size
img_size = 64

# Empty list to store the training images after conversion to Numpy arrays
train_images = []

# Empty list to store the training labels
train_labels = []
```

```python
# We will apply the same procedure to the "parasitized" and "uninfected" folders within the "train" directory.
for directory_name in ['/parasitized/', '/uninfected/']:

    # Path of the folder
    images_path = os.listdir(train_dir + directory_name)

    for i, image_name in enumerate(images_path):

        try:

            # Opening each image using the path of that image
            image = Image.open(train_dir + directory_name + image_name)

            # Resizing each image to (64, 64)
            image = image.resize((img_size, img_size))

            # Converting images to arrays and appending that array to the empty list defined above
            train_images.append(np.array(image))

            # Creating labels for parasitized and uninfected images
            if directory_name == '/parasitized/':

                train_labels.append(1)

            else:

                train_labels.append(0)

        except Exception:

            pass

# Converting lists to arrays
train_images = np.array(train_images)

train_labels = np.array(train_labels)
```

```python
# Creating a variable to store the path of the extracted train dataset
test_dir = '/content/drive/MyDrive/Learning/ADSP/Capstone_Project/datasets/cell_images/test'

# Designed size so all the images have the same size
img_size = 64

# Empty list to store the test images after conversion to Numpy arrays
test_images = []

# Empty list to store the test labels
test_labels = []
```

```python
# We will run the same code for "parasitized" as well as "uninfected" folders within the "test" folder
for directory_name in ['/parasitized/', '/uninfected/']:

    # Path of the folder
    images_path = os.listdir(test_dir + directory_name)

    for i, image_name in enumerate(images_path):

        try:

            # Opening each image using the path of that image
            image = Image.open(test_dir + directory_name + image_name)

            # Resizing each image to (64, 64)
            image = image.resize((img_size, img_size))

            # Converting images to arrays and appending that array to the empty list defined above
            test_images.append(np.array(image))

            # Creating labels for parasitized and uninfected images
            if directory_name == '/parasitized/':

                test_labels.append(1)

            else:

                test_labels.append(0)

        except Exception:

            pass

# Converting lists to arrays
test_images = np.array(test_images)

test_labels = np.array(test_labels)
```

###<b> Processing Images using Gaussian Blurring

###<b> Building the model

We saw that the ANN, while versatile, struggled with spatial data, whereas the CNN excelled in image classification due to its ability to capture spatial features effectively (cf. Elective project - Deep learning). For that, we will start the modeling with a CNN.

```python
# Fixing the seed

np.random.seed(42)

import random

random.seed(42)

tf.random.set_seed(42)
```

```python
# Let's create a convolutional neuron network

# Define the CNN model
model1 = Sequential()

# Convolutional Layer 1
model1.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', activation = 'relu', input_shape = (64, 64, 3)))
model1.add(MaxPooling2D(pool_size = (2, 2)))
model1.add(Dropout(0.2))

# Convolutional Layer 2
model1.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', activation = 'relu', input_shape = (64, 64, 3)))
model1.add(MaxPooling2D(pool_size = (2, 2)))
model1.add(Dropout(0.2))

# Flatten the layers
model1.add(Flatten())

# Fully Connected Layer 1
model1.add(Dense(units = 128, activation = 'relu'))
model1.add(Dropout(0.2))

# Fully Connected Layer 2
model1.add(Dense(units = 64, activation = 'relu'))
model1.add(Dropout(0.2))

# Output Layer
model1.add(Dense(units = 2, activation = 'softmax'))

# Model summary
model1.summary()
```

```python
# Function to plot train and validation accuracy
def plot_accuracy(history):

    N = len(history.history["accuracy"])

    plt.figure(figsize = (7, 7))

    plt.plot(np.arange(0, N), history.history["accuracy"], label = "train_accuracy", ls = '--')

    plt.plot(np.arange(0, N), history.history["val_accuracy"], label = "val_accuracy", ls = '--')

    plt.title("Accuracy vs Epoch")

    plt.xlabel("Epochs")

    plt.ylabel("Accuracy")

    plt.legend(loc="upper left")
```

```python
from tensorflow.keras.applications.vgg16 import VGG16

from tensorflow.keras import Model

vgg_model = VGG16(include_top = False, weights = 'imagenet', input_shape = (64, 64, 3))

vgg_model.summary()
```

```python
# Getting only the Conv layers for transfer learning
transfer_layer = vgg_model.get_layer('block5_pool')

# Making all the layers of the VGG model non-trainable. i.e. freezing them
for layer in vgg_model.layers:
    layer.trainable = False


x = Flatten()(transfer_layer.output)
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
x = BatchNormalization()(x)
prediction = Dense(2, activation='softmax')(x)

vgg_model = Model(inputs = vgg_model.input, outputs = prediction)

vgg_model.summary()
```

## <b>Model with HSV images

```python
# Clearing the previous model history
from tensorflow.keras import backend
backend.clear_session()

# Fixing the seed
np.random.seed(42)
import random
random.seed(42)
tf.random.set_seed(42)
```

###<b>Building the Model

```python
# Define the new CNN
model_hsv = Sequential()

# Convolutional Layer 1
model_hsv.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', input_shape = (64, 64, 3)))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(MaxPooling2D(pool_size = (2, 2)))
model_hsv.add(Dropout(0.2))

# Convolutional Layer 2
model_hsv.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', input_shape = (64, 64, 3)))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(MaxPooling2D(pool_size = (2, 2)))
model_hsv.add(Dropout(0.2))

# Convolutional Layer 3
model_hsv.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', input_shape = (64, 64, 3)))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(MaxPooling2D(pool_size = (2, 2)))
model_hsv.add(Dropout(0.2))

# Convolutional Layer 4
model_hsv.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', input_shape = (64, 64, 3)))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(MaxPooling2D(pool_size = (2, 2)))
model_hsv.add(Dropout(0.2))

# Convolutional Layer 5
model_hsv.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', input_shape = (64, 64, 3)))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(MaxPooling2D(pool_size = (2, 2)))
model_hsv.add(Dropout(0.2))

# Flatten the layers
model_hsv.add(Flatten())

# Fully Connected Layer 1
model_hsv.add(Dense(units = 512))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(Dropout(0.2))

# Fully Connected Layer 2
model_hsv.add(Dense(units = 128))
model_hsv.add(BatchNormalization())
model_hsv.add(LeakyReLU(alpha=0.1))
model_hsv.add(Dropout(0.2))

# Output Layer
model_hsv.add(Dense(units = 2, activation = 'softmax'))

# Model summary
model_hsv.summary()
```
